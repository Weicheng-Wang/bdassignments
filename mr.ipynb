```python
df = pd.read_csv("HTTP_20130313143750.dat", sep='\t', header=None, names=["时间戳","电话号码", "基站的物理地址", "访问网址的ip", "网站域名", "网站类型", "数据包", "接包数", "上行/传流量", "下行/载流量", "响应码"]) 
```

```python
df.head(25)
```

```python
aggdf = df.groupby("电话号码", as_index=False)[["上行/传流量", "下行/载流量"]].sum()
```

```python
aggdf["总流量"] = aggdf[["上行/传流量", "下行/载流量"]].sum(axis=1)
```

```python
aggdf.head()
```

```python
aggdf = aggdf.sort_values("电话号码")
```

```python
def sortkey(col):
  if (col // 1000000000) == 0:
    return col * 1000
  else:
    return col
```

```python
84138413 // 1000000000
```

```python
aggdf["sort"] = aggdf["电话号码"].apply(sortkey)
```

```python
aggdf = aggdf.sort_values("sort").drop("sort",  axis=1)
```
